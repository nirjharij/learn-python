<h1>Variables in Python</h1>

 - Variables are used for storing data in Python
 - Can store different types in variables like Numbers - integer & float, Text and boolean

```python
# Storing integers in variables
a = 10
type(a)
```

```python
# Storing text in variables
b="This is a text"
type(b)
```

```python
# Storing boolean in variables
c=True
type(c)
```

```python
# Address location of variables
x=5
print("Address of variable X: " + str(id(x)))
y=10
print("Address of variable Y: " + str(id(y)))
z=x
print("Address of variable Z: " + str(id(z)))
```
