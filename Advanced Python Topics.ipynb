Packing and Unpacking Arguments:


```python
#Example of unpacking of arguments
mylist = [1,2,4,5]
print(*mylist)
args = [3,6]
range(*args)
```

```python
#Example of packing of arguments
def sum(*args):
    s = 0
    for i in range(0,len(args)):
        s = s+args[i]
    return s

sum(1,2,3,4,5)
```

```python
#Example of unpacking dictionary
def mul(a,b,c):
    return a*b*c
d = {"a":1,"b":2,"c":3}
mul(**d)
```

Here ** unpacked the dictionary used with it, and passed the items in the dictionary as keyword arguments to the function. So writing “mul(1, **d)” was equivalent to writing “mul(1, b=2, c=3)”.



Partial Functions in Python:

Partial functions allow us to fix a certain number of arguments of a function and generate a new function.

```python
from functools import partial

def func(a,b,c,d):
    return a+b+c+d

partial_func = partial(func,1,2)
partial_func(3,4)
```

In the example we have pre-filled our function with some constant values of a and b. And partial_func() just takes two arguments i.e c and d.
