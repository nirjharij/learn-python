```python
s = 'Hello World'
print (str(s))
```

```python
s = 'Hello World'
print (repr(s))
```

str is mainly used for creating output for end user while repr is mainly used for debugging and development.
print statement and str builtin function uses __str__ to display the string representation of the object while repr uses __repr__ to display the object.

```python
import datetime
today = datetime.datetime.now()
print (str(today))
print (repr(today))
today
```

str() displays today’s date in a way that the user can understand the date and time while repr() prints “official” representation of a date-time object

A user defined class should also have a __repr__ if we need detailed information for debugging. 

```python
class Complex:
    def __init__(self,real,imag):
        self.real = real
        self.imag = imag
    def __str__(self):
        return "%s + i%s" %(self.real,self.imag)
    def __repr__(self):
        return "Rational(%s,%s)" %(self.real,self.imag)  
```

```python
obj = Complex(1,5)
print (str(obj))
print (repr(obj))
```

In Python Functions can be considered “call by reference” . If the object allows updating then it behaves as you would expect from “call by reference”, but if object’s value cannot be changed, then it will behave like “call by value”. Example:

```python
def call_by_ref(list_obj):
    list_obj.append(2)

a_list = [1,3,4,5]    
call_by_ref(a_list)
print(a_list)

def call_by_val(tuple_obj):
    print("id inside call by value without altering:",id(tuple_obj))
    tuple_obj = (4,3,2,1)
    print("id inside call by value after altering:",id(tuple_obj))
    tuple_obj = tuple_obj+(5,)
    print("id inside call by value after adding an element:",id(tuple_obj))
    print ("Final tuple inside call by value",tuple_obj)
    
a_tuple = (1,2,3,4)
call_by_val(a_tuple)
print("outside function definition",a_tuple)
```

Python Memory Management:

Python keeps track of all the objects in use by reference counting and when the count for a particular object becomes zero that object is garbage collected.
The ref count for an object is initially set to 1 when an object is created and assigned. New references to objects also called aliases, occur when additional variables are assigned to the same object, passed as arguments to invoke other bodies of code such as functions, methods or class instantiation, or assigned as members of a sequence or mapping.

Objects ref count is increased when:
1. It is created x = 3.14
2. Additional aliases are created y = x
3. Passed to a function foo(x)
4. Becomes part of container object mylist = [123, x, ‘xyz’]

Reference count is decreased when a reference goes out of scope(ie when the function in which a reference is made completes)
1. A reference also goes away when a variable is reassigned to another object. 
2. Explicit removal of reference using del statement
3. Del removes a single reference to an object.
Memory that is unused is reclaimed by the system using a mechanism called garbage collection

Python additionally has a cyclic garbage collector also which clears all the cyclic referenced objects i.e. when at least two objects that refer to each other.

gc.set_threshold(threshold0[, threshold1[, threshold2]])
Set the garbage collection thresholds (the collection frequency). Setting threshold0 to zero disables collection.

The GC classifies objects into three generations depending on how many collection sweeps they have survived. New objects are placed in the youngest generation (generation 0). If an object survives a collection it is moved into the next older generation. Since generation 2 is the oldest generation, objects in that generation remain there after a collection. In order to decide when to run, the collector keeps track of the number object allocations and deallocations since the last collection. When the number of allocations minus the number of deallocations exceeds threshold0, collection starts. 

```python
import gc
print(dir(gc))
print("Garbage collection thresholds:",gc.get_threshold())
print(gc.get_count())
gc.collect()
```
